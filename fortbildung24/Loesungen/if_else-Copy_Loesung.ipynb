<h1>if-Bedingung in Python</h1>

<p>In Python gibt es die Möglichkeiten Bedingungen zu überprüfen und entsprechend im Programmablauf darauf zu reagieren. Hier könnten wir abhängig von der Uhrzeit den Nutzer entsprechend Begrüßen. Aber erst einmal der allgemeine Aufbau von <code>if</code>-Abfragen und wie wird diese einsetzen.</p>


<h2>Aufbau der <code>if</code>-Abfrage</h2>

<p>Würden wir in Deutsch eine Bedingung formulieren, würde diese wie folgt aussehen</p>

<div class="browserfenster">

<p>wenn Wert kleiner als 5 dann mache alles Doppelpunkt</p>
</div>


<p>Und nun das Ganze als Python-Programm, in der wir als Erstes die Variable definieren:</p>

<pre><code class="language-python">wert = 3
if wert &#60; 5:
    print('Wert ist kleiner als 5')
</code></pre>


<p>Das wichtige ist, dass nach der <code>if</code>-Abfrage das weitere, was zu der <code>if</code>-Abfrage gehört, <strong>eingerückt wird</strong>! Diese Einrückungen sind der Dreh- und Angelpunkt in Python und garantiert einen sauberen Quellcode. Pfuscht man bei den Einrückungen, funktioniert das Programm nicht wie erwartet.</p>

<p>Bei unserer <code>if</code>-Anfrage gehört alles Folgende, was eingerückt ist, zu der Bedingung. Und das kann mehr als eine Zeile sein!</p>

<pre><code class="language-python">wert = 3
if wert &#60; 5:
    print('Wert ist kleiner als 5')
    print('Ich gehöre auch noch zu der Bedingung')
    
print('und hier geht es nach der if-Abfrage weiter')
</code></pre>


<p>Lassen wir das Programm ausführen, erhalten wir als Ergebnis alle 3 Zeilen ausgegeben. Die 2 Textzeilen, die „innerhalb“ der <code>if</code>-Abfrage sind (sprich, die eingerückt sind) und die letzte Textzeile, die nach der <code>if</code>-Abfrage kommt und nicht eingerückt ist.</p>

<p>Ändern wir nun unsere Variable am Anfang auf <code>wert = 9"</code>, erhalten wir bei der Programmausführung nur noch die letzte Zeile ausgegeben, die nach der <code>if</code>-Abfrage kommt. Die <code>if</code>-Abfrage ist nicht wahr, weil der Wert mit 9 bereits größer ist als die Bedingung <code>&#60; 5</code> und somit das eingerückte der <code>if</code>-Abfrage nicht ausgeführt wird.</p>

<p>Wichtig ist bei Programmänderungen, dass man speichern nicht vergisst!</p>

<p>Was passiert eigentlich genau bei <code>wert &#60; 5</code>? Python überprüft, ob das Ergebnis wahr oder falsch ist. Dabei kann auch direkt „true“ oder „false“ der <code>if</code>-Abfrage präsentiert werden und diese reagiert darauf entsprechend:</p>

<pre><code class="language-python">if True:
    print('if-Bedingung ist wahr')
</code></pre>


<p>Es erscheint als Ergebnis:</p>

<div class="browserfenster">

<p>if-Bedingung ist wahr</p>
</div>


<p>Natürlich könnten wir auch eingeben <code>if false:</code> und es würde nichts angezeigt werden, da unsere <code>if</code>-Abfrage nicht wahr ist.</p>

<p>Natürlich würde kein Mensch so eine <code>if</code>-Abfrage erstellen, denn die <code>if</code>-Abfrage würde ja immer exakt zum gleichen Ergebnis führen. Allerdings können wir auch Variablen übergeben, die diesen Wert haben.</p>

<pre><code class="language-python">wahroderfalsch = True
if wahroderfalsch:
    print('if-Bedingung ist wahr')
</code></pre>


<p>Unsere Variable "wahroderfalsch" hat nun den Wert "true" zugewiesen bekommen und die <code>if</code>-Abfrage reagiert entsprechend darauf.</p>

<p>Diese Variablen sind vom Typ Boolean – es gibt nur die Werte <code>true</code> oder <code>false</code> bei Boolean. Diese Variablen-Typen sind benannt nach dem Erfinder George Boole.</p>


<h2>Ungleich <code>!=</code> in Python</h2>

<p>Oft möchte man auch einfach wissen, ob eine bestimmte Bedingung nicht zutrifft, also ungleich ist.</p>

<p>Dazu kann man den Operator <code>!=</code> (ungleich) nutzen.</p>

<p>Als Beispiel wollen wir von einer Stundenzahl wissen, ob es NICHT 12 Uhr ist. Dazu unser Code:</p>

<pre><code class="language-python">wert = 11
if wert != 5: 
    print('Es ist nicht 12 Uhr')
</code></pre>



<h2>Alternative, wenn <code>if</code>-Abfrage nicht zutrifft</h2>

<p>Jetzt ist es oft geschickt gleich darauf reagieren zu können, wenn die <code>if</code>-Abfrage nicht zutrifft. Ohne diese Kenntnis müssten wir solche kantigen Konstruktionen bauen wie folgt:</p>

<pre><code class="language-python">wert = 9
if wert &#60; 5:
    print('Wert ist kleiner als 5')
    
if wert &#62; 4:
    print('Wert ist größer als 4')
</code></pre>


<p>Dies wollen wir kürzer, da die zweite <code>if</code>-Abfrage einfach das Gegenteil gerade von der ersten darstellt. Alles was nicht kleiner ist als 5 ist auf jeden Fall größer als 4. Und hier kommt das schöne Wort <code>else</code> zum Einsatz:</p>

<pre><code class="language-python">wert = 9
if wert &#60; 5:
    print('Wert ist kleiner als 5')
else:
    print('Wert ist größer als 4')
</code></pre>


<p>Unseren zweiten Part ersetzen wird durch <code>else:</code>. Trifft unsere <code>if</code>-Abfrage nicht zu, sprich ist diese nicht wahr, sondern falsch, dann wird der Block unter <code>else:</code> ausgegeben:</p>

<p>Als Ausgabe erhalten wir nun bei der Programmausführung:</p>

<p>Wert ist größer als 4</p>

<h2>Weitere Bedingungen innerhalb der Bedingung prüfen – elif</h2>

<p>Unser bisheriges Programm ist nicht wirklich sexy. Es gibt die Ausgabe „Wert ist kleiner als 5“ oder „Wert ist größer als 4“. Eigentlich wären folgenden 3 mögliche Ergebnisse deutlich schicker.</p>

<ul>
  <li>Wert ist kleiner 5</li>
  <li>Wert ist exakt 5</li>
  <li>Wert ist größer 5</li>
</ul>

<p>Dazu brauchen wir eine Abfrage innerhalb der Abfrage. Und dazu kennt Python den Befehl <code>elif</code>. In den meisten anderen Programmiersprachen kennt man dies als „elseif“ aber in Python ist es <code>elif</code>.</p>

<p>Unser Programm von oben sieht nun damit so aus.</p>

<pre><code class="language-python">wert = 9
if wert &#60; 5:
    print('Wert ist kleiner als 5')
elif wert == 5:
    print('Wert ist exakt 5')
else:
    print('Wert ist größer als 5')
</code></pre>


<p>Der Überprüfung von der Bedingung hinter <code>elif</code> wird nur dann überprüft, wenn die erste Bedingung nicht zutrifft. Trifft unsere Bedingung unter <code>elif</code> auch nicht zu, dann kommt die Alternative unter <code>else</code> zum Tragen.</p>


<p>Wichtig sind die Einrückungen!</p>

<h2>Kleine böse Fehlerquelle – doppeltes Gleichzeichen</h2>

<p>
Hier unser Code, in dem sich ein Fehler eingeschlichen hat:
</p>

<pre><code class="language-python">wert = 9
if wert &#60; 5:
    print('Wert ist kleiner als 5')
elif wert = 5:
    print('Wert ist exakt 5')
else:
    print('Wert ist größer als 5')
</code></pre>


<p>Wir überprüfen bei <code>elif</code>, ob dieses exakt 5 entspricht. Hier ist der Standardfehler, dass versehentlich nur ein Gleichzeichen (anstelle von 2) gemacht wurde. Das ist aber nicht möglich und somit kommt eine Fehlermeldung von Python in Form von „SyntaxError: invalid syntax“</p>

**1.0 Rabattrechner**: Eingabe soll der Normalpreis sein und wie viel Prozent Rabatt es auf den Artikel gibt. Ausgabe soll der Endpreis sein. Weise den Nutzer darauf hin, wenn er 100% oder mehr eingibt!

**2.0 quadratische Gleichung** Das Programm soll eine quadratische Gleichung der Form $$ax^2 + bx + c = 0$$
 lösen können. Der Nutzer gibt hierfür die Zahlen a,b,c ein. Das Programm prüft die Diskriminante, also$$b^2 - 4ac$$
  ob diese positiv, null oder negativ ist und gibt dann aus wie viele Lösungen es gibt und ggf. deren Werte.
