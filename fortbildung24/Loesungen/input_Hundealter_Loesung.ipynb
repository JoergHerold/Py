## 1.3 Die input-Funktion (Interaktion mit dem Nutzer)

<h2>Eingaben vom Nutzer abfragen in Python</h2>

<p>Wir haben bei den vorherigen Aufgaben immer wieder Daten auf dem Bildschirm ausgegeben. Allerdings werden in der Realität die Daten entweder aus einer Datenbank kommen oder durch einen Nutzer festgelegt, also eingegeben werden.Um eine Angabe vom Nutzer anzufordern, gibt es in Python den Befehl <code>input</code></p>

<p>Diesem Befehl können wir einen Text mitgeben, damit dem Nutzer klar ist, was er denn eingeben soll. Die Eingabe selber wird in einer <b>Variablen</b> gespeichert, mit der wir dann weiterarbeiten können. Unser Befehl sieht dann so aus:</p>

<pre><code class="language-python">benutzereingabe = input("Bitte Zahl eingeben")
</code></pre>


<p>Das Python-Programm pausiert nun, wenn es diese Programmzeile erreicht hat und wartet ab, bis der Nutzer eine Eingabe gemacht hat. Erst nach der Eingabe läuft das Programm dann weiter. Nun können wir die Variable, die im Beispiel „benutzereingabe“ genannt wurde, im weiteren Programmablauf nutzen.</p>

<pre><code class="language-python">benutzereingabe = input("Bitte Zahl eingeben")
print(benutzereingabe)
</code></pre>


<h2>Input liefert Strings</h2>

<p>Als Rückgabewert der <code>input</code>-Funktion erhalten wir eine String. Den <b>Typ einer Variablen</b> können wir uns über die Funktion <code>type</code> ausgeben lassen. Im folgenden Beispiel zum Testen:</p>

<pre><code class="language-python">benutzereingabe = input("Bitte Zahl eingeben")
print(type(benutzereingabe))
</code></pre>


<p>Als Ergebnis erhalten wir:

<div class="browserfenster">

class 'str'
</p></div>


<p>Wollen wir aber mit der vom Benutzer eingegebenen Zahl weiterrechnen, müssen wir den String erst in eine Zahl konvertieren. <b>Der Datentyp einer Variablen wird umgewandelt</b>. Wenn man den Typ einer Variablen umwandelt, spricht man auch von „casting“.  Das Wort erinnert nicht zu Unrecht an das Besetzen von Filmrollen (also die Rollenverteilung). Und genau das machen wir mit dem Typ. Wir sagen, du bist nun eine Ganzzahl (integer).</p>

<pre><code class="language-python">benutzereingabe = int(benutzereingabe)
</code></pre>


<p>Jetzt können wir mit der Benutzereingabe als Zahl arbeiten. Diese Zahl können wir über <code>print</code> ausgeben lassen.</p>

<pre><code class="language-python">print(benutzereingabe)
</code></pre>

**Situation:** Sie machen Praktikum im Tierheim Bamberg und sollen sich dort um die Digitalisierung kümmern. Ihr erster Auftrag ist es, eine Abfrage für die Neuanmeldung eines Tieres anzulegen. Vorab: Der Fragebogen soll Name des Hundes und das Alter abfragen. Zusätzlich soll auch berechnet werden können, wie alt der Hund in Menschenjahren ist. 

**Arbeitsauftrag:** Schreiben Sie den Programmcode mit Hilfe der untenstehenden Pseudocode-Anweisungen


$$  \text{Hundealter in Menschenjahren} = \text{Hundealter} \times 7  $$

````python
hundename = input("Geben Sie den Hundenamen ein: ")
hundealter = int(input("Geben Sie das Hundealter ein: "))
hundeAlterinMenschenJahren = hundealter*7
print(hundename,"ist", hundeAlterinMenschenJahren, "Jahre alt.")
````
