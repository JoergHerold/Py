## 4 Einkaufskalkulation

Der Benutzer soll zu Beginn den Listeneinkaufspreis und die Bezugskosten eingeben. Mit Hilfe von Python soll dann der Bezugspreis berechnet werden. Der Rabatt beträgt 10%, der Skonto 2%.


```python
lep=input("Bitte geben Sie den Listeneinkaufspreis an: ")
bk=input("Bitte geben Sie die Bezugskosten an: ")

lep=float(lep)
bk=float(bk)

lep=round(lep,2)
bk=round(bk,2)

rabatt=0.1
skonto=0.02

rabatt=round(rabatt*lep,2)
zep=round(lep-rabatt,2)
skonto=round(zep*skonto,2)
bep=zep-skonto
bezugspreis=round(bep+bk,2)

print("Einkaufskalkulation")
print("===================================")
print("  Listeneinkaufspreis", format(lep,'.2f'))
print("- Lieferrabatt       ", format(rabatt,'.2f'))
print("___________________________________")
print("  Zieleinkaufspreis: ",format(zep,'.2f'))
print("- Skonto:            ",format(skonto,'.2f'))
print("___________________________________")
print("  Bareinkaufspreis:  ", format(bep,'.2f'))
print("+ Bezugskosten:      ",format(bk,'.2f'))
print("___________________________________")
print("  Bezugspreis:       ",format(bezugspreis,'.2f'))
```

# Zusatzaufgabe

Schon fertig? Dann erweitern Sie die Kalkulation bitte noch um eine Eingabe für Skonto und Rabatt sowie um die Umsatzsteuer.

```python
lep=int(input("Bitte geben Sie den Listeneinkaufspreis an:"))
bk=int(input("Bitte geben Sie die Bezugskosten ein:"))
lr=int(input("Bitte geben Sie den Rabatt in Prozent ein:"))
skonto=int(input("Bitte geben Sie das Skonto in Prozent ein:"))

lr=lep*(lr/100)
zep=lep-lr
skonto=zep*(skonto/100)
bep=zep-skonto
bp=bep+bk

print("Listeinkaufspreis:",lep)
print("-Lieferrabatt:",lr)
print("=Zieleinkaufspreis:",zep)
print("-Skonto:",skonto)
print("Bareinkaufspreis:",bep)
print("+Bezugskosten:",bk)
print("=Bezugspreis:",bp)
print("=Bezugspreis inkl Ust.:",bp*119/100)
```
